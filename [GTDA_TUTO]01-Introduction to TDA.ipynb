# 1. Introduction to TDA.

__Topological Data Analysis(TDA)__ 는 데이터의 모양(shape)이라는 정성적 특징(qualitative feature)을 위상수학을 이용해 추출하고 사용하는 수학적 데이터 분석 방법론입니다. 

데이터의 모양은 데이터의 종류에 따라 여러 형태로 해석될 수 있습니다.

예를 들어, 좌표공간 상에 흩어진 유한 개 점들의 데이터가 있다면 이 점들이 대략적으로 띄는 모양을 상상할 수 있을 것입니다.

이는 마치 무작위로 흩어진 듯 보이는 밤하늘의 별들로부터 몇 개의 점들을 이어 북극곰이나 전갈 따위의 형상을 생각하는 별자리의 아이디어와 비슷합니다.

또 다른 예로는 손글씨로 쓰여진 이미지 데이터가 있다면, 숫자들에 구멍이 몇 개인지와 같은 특징을 상상할 수도 있을 것입니다.

이런 정보를 어떻게 하면 데이터로부터 추출할 수 있을까요? 

이런 정보를 기계 학습에 사용하려면 또 어떻게 해야 할까요?

TDA는 위와 같은 질문에 대해 수학적으로 합리적인 대답을 제시합니다.

그리고 이런 정성적 정보를 기계 학습에 적합한 정량적 특징으로 변환해 사용할 수 있는 방법도 제시합니다.

## 3.1. Point cloud

그러나 누군가에겐 이런 무작위성 중에 어떤 기하학적 특징이 보일 수도 있습니다.

## 3.2 Vietoris-Rips Complex

아래 그림과 같이 유클리드 공간 상에  8개의 점으로 이루어진 포인트 클라우드가 주어졌다고 합시다.

```python
## Generate a point-cloud with 8 points.
import numpy as np
import matplotlib.pyplot as plt

point_cloud = np.array([[-2.0,8.0],
                        [6.0,8.0],
                        [-2.0,2.0],
                        [7.0,3.0],
                        [27.0, 3.0],
                        [20.0,6.0],
                        [20.0,0.0],
                        [9.0,-7.0]
                        ])

plt.plot(point_cloud[:,0], point_cloud[:,1], '.')
```

위 코드를 보시면 `matplotlib`으로 포인트 클라우드의 플롯을 보려면 조금은 신경써야할 부분들이 있고, 그림이 세련되지도 않습니다.

`giotto-tda`는 포인트 클라우드를 입력하는 것 만으로도 손쉽게 세련된 포인트 클라우드 그림을 보여주는 기능을 가지고 있습니다.

이는 `plotly`라는 라이브러리의 기능을 이용하는 것이며, 포인트 위에 마우스 포인터를 올리면 점의 좌표를 살펴볼 수 있고, 포인트 클라우드를 확대 및 축소할 수도 있으며, 3차원 포인트클라우드의 경우 회전 시켜 보는 등 다양한 기능을 쉽게 사용할 수 있습니다.

```python
## plotting point cloud through giotto-tda.
from gtda.plotting import plot_point_cloud
plot_point_cloud(point_cloud)
```

여러분은 위 점들에서 어떤 기하학적 직관을 얻을 수 있나요? 

저는 왼쪽 상단의 네 개의 점이 마치 하나의 사각형을 이룰 것 같은 느낌을 받습니다.

여러분도 어쩌면 머릿속에서 자연스럽게 *서로 가까운 두 점을 선분으로 이어서* 어떠한 기하학적 모양을 추측하려 했을 것입니다.

이런 단순한 아이디어를 가지고 포인트 클라우드로부터 위상적인 공간을 근사적으로 구성한 것을 *Vietoris-Rips Complex*라고 합니다.

조금 더 형식적으로 설명하자면 아래와 같은 과정을 통해 하나의 위상공간을 구성해 나갑니다.

1. 각 점을 중심으로 반지름이 $\epsilon$인 원을 생각합니다.

2. 두 점 $i$와 $j$ 사이의 거리보다 $\epsilon$이 큰 값을 가지면, 즉 각 점을 중심으로 하고 반지름이 $\epsilon$인 두 원이 서로의 중심을 포함하며 겹치면 두 점 사이에 변을 하나 그립니다.

3. 세 변이 하나의 삼각형을 이루면, 그 면을 채웁니다.

아래에 $\epsilon = 8$일 때 Vietoris-Rips complex를 시각화하는 코드를 작성해두었습니다.

```python
## Visualize Vietoris-Rips Complex

from matplotlib.patches import Circle,Polygon

epsilon = 8     # distance parameter
m = len(point_cloud)    # the number of points in given point cloud
adj_mat = np.zeros((m,m))

## Assume that the metric for our point cloud is Euclidean distance.
def Euclidean_distance(x,y):
    norm = np.dot(x-y,x-y)
    return np.sqrt(norm)

plt.figure(figsize=(10,10))
## Step 1. For each point p, draw a circle with radius epsilon and center p
plt.plot(point_cloud[:,0], point_cloud[:,1], '.')
for i in range(0,len(point_cloud)):
    plt.gca().add_patch(Circle(point_cloud[i], radius = epsilon, alpha = 0.3))
    plt.annotate(f'{i+1}',(point_cloud[i]))


## Step 2. If the distance between two points x and y is less than epsilon, connect two points with edge.
for x,i in zip(point_cloud,np.arange(m)):
    for y,j in zip(point_cloud,np.arange(m)):
        if 0 < Euclidean_distance(x,y) < epsilon:
            plt.plot([x[0],y[0]],[x[1],y[1]],'-')
            adj_mat[i,j] = 1
            adj_mat[j,i] = 1


## Step 3. If three edges consist of a triangle, fill in the face of the triangle.
for i in range(0,m):
    for j in range(0,m):
        for k in range(0,m):
            if (adj_mat[i,j] == 1) & (adj_mat[j,k] == 1) & (adj_mat[k,i] == 1):
                plt.gca().add_patch(Polygon((point_cloud[i],point_cloud[j],point_cloud[k]), facecolor = 'm'))

plt.title(f'Vietoris-Rips complex of point cloud for $\epsilon$ = {epsilon}', fontsize = 14)
plt.xlim(np.min(point_cloud)-epsilon,np.max(point_cloud)+epsilon)
plt.ylim(np.min(point_cloud)-epsilon,np.max(point_cloud)+epsilon)
plt.grid()
```

위 코드를 함수로 작성해 다음에도 이용할 수 있도록 합시다.

```python
## Visualize Vietoris-Rips Complex

from matplotlib.patches import Circle,Polygon



def Visualize_VRC(point_cloud, epsilon):
    epsilon = epsilon     # distance parameter
    m = len(point_cloud)    # the number of points in given point cloud
    adj_mat = np.zeros((m,m))

    ## Assume that the metric for our point cloud is Euclidean distance.

    
    plt.figure(figsize=(10,10))
    ## Step 1. For each point p, draw a circle with radius epsilon and center p
    plt.plot(point_cloud[:,0], point_cloud[:,1], '.')
    for i in range(0,len(point_cloud)):
        plt.gca().add_patch(Circle(point_cloud[i], radius = epsilon, alpha = 0.3))
        plt.annotate(f'{i+1}',(point_cloud[i]))


    ## Step 2. If the distance between two points x and y is less than epsilon, connect two points with edge.
    for x,i in zip(point_cloud,np.arange(m)):
        for y,j in zip(point_cloud[i:],np.arange(i,m)):
            if 0 < Euclidean_distance(x,y) < epsilon:
                plt.plot([x[0],y[0]],[x[1],y[1]],'-')
                adj_mat[i,j] = 1
                adj_mat[j,i] = 1


    ## Step 3. If three edges consist of a triangle, fill in the face of the triangle.
    for i in range(0,m):
        for j in range(0,m):
            for k in range(0,m):
                if (adj_mat[i,j] == 1) & (adj_mat[j,k] == 1) & (adj_mat[k,i] == 1):
                    plt.gca().add_patch(Polygon((point_cloud[i],point_cloud[j],point_cloud[k]), facecolor = 'm'))
    
    plt.title(f'Vietoris-Rips complex of point cloud for $\epsilon$ = {epsilon}', fontsize = 14)
    plt.xlim(np.min(point_cloud)-1,np.max(point_cloud)+1)
    plt.ylim(np.min(point_cloud)-1,np.max(point_cloud)+1)
    plt.grid()
```

이 함수를 이용해 $\epsilon$을 0에서부터 점점 증가시켜 나가면서 Vietoris-Rips Complex의 모양이 어떻게 변해가는지 살펴보기 쉬워졌습니다.

아래 코드의 결과로 나오는 영상을 보면서 포인트 클라우드가 기하학적으로 어떤 모양을 가지는지 생각해보세요.

```python
from pylab import *
%matplotlib inline

e = np.linspace(0,19,20)

for e in e:
    Visualize_VRC(point_cloud, e)
```

$\epsilon$ 값이 증가함에 따라, 점들은 변으로 이어지게 되고 그러면서 우측에 삼각형이 보이기도 하고 좌측에 사각형이 보이기도 하며 종국에는 하나의 다각형으로 뭉쳐지게 됩니다.

여기서 $\epsilon$ 값이 10 근방일 때를 주목해 봅시다. 

이때의 Vietoris-Rips Complex의 그림을 살펴보면, 왼쪽 사각형은 채워진 도형이 아니므로 하나의 구멍이 있는 것처럼 보이는 것을 알 수 있습니다.

위 관찰을 통해서 우리가 살펴볼 수 있는 것은 $\epsilon$이 증가함에 꼭짓점, 변, 삼각형들이 새롭게 탄생하거나 서로 뭉쳐지면서 하나의 연결 성분이 되어 그 총 개수는 줄어드는 것을 살펴볼 수 있습니다.

또한 때로는 변들이 어떤 구멍을 형성하기도 하고, 이러한 구멍이 다른 변들로 채워져서 사라지기도 하는 것을 살펴볼 수 있습니다.

이처럼 *연결 성분의 수*, *구멍의 수* 등은 어떤 위상공간이 가지는 기하학적 특징을 나타내 주는 중요한 지표가 될 수 있습니다.

이러한 정보는 Complex의 homology를 계산하는 것에 의해 구해질 수 있습니다.


## 3.3 Persistent homology
그런데, 고정된 $\epsilon$에 대해서 하나의 Vietoris-Rips Complex를 생각하고자 한다면, 어떤 $\epsilon$이 데이터의 모양을 가장 잘 나타내는 것인지 모호합니다.

그러므로 우리는 모든 $\epsilon$에 대해 연결 성분의 수와 구멍의 수 등이 어떻게 변화하는지 일괄적으로 요약할 필요가 있습니다.

TDA의 주요 도구인 __Persistent homology__ 는 이러한 생각에서 도입된 개념입니다.

위와 같이 파라미터 $\epsilon$에 의해 변화해가는 Complex들로부터 어떤 homology에 대한 정보가 탄생하고 소멸하는지, 즉 얼마나 지속하는지(persist)를 측정하는 개념이 바로 __Persistent homology__ 입니다.


### Persistent barcode & Persistent Diagram

위 예제에서 위상적인 정보가 $\epsilon$의 변화에 따라 어떻게 변해가는지 찬찬히 살펴봅시다.

- $\epsilon$이 $[0,7)$에 있는 동안에 Complex에는 단지 8개의 꼭짓점들만 있을 뿐입니다.
    다시 말해 이 구간에서 Complex의 **연결성분**은 **8**개 입니다.


- 그런데 $\epsilon = 7$이 되는 순간, 꼭짓점 1과 3, 2와 4, 6과 7이 변으로 이어지게 됩니다.
    따라서 Complex에는 꼭짓점이 2개, 변이 3개로 총 **5**개의 **연결성분**이 존재합니다.


- $\epsilon=8$이 되면 꼭짓점 2와 5, 4와 5가 변으로 이어지면서 하나의 삼각형으로 합쳐집니다.
    그러므로 이때 Complex에는 꼭짓점 1개, 변이 2개, 삼각형이 1개로 총 4개의 **연결성분**이 존재합니다.


- $\epsilon = 9$가 되면 꼭짓점 1과 2가 하나의 변으로 이어집니다.
    그러므로 세 개의 변이 하나의 부분적으로 매끄러운 연결성분으로 이어집니다.
    따라서 Complex에는 총 **3**개의 **연결성분**이 존재합니다.


- $\epsilon = 10$이 되면 꼭짓점 3과 4가 연결되면서 하나의 *구멍이 뚫린* 사각형이 만들어 집니다.
    따라서 Complex에는 **3**개의 **연결성분**과 **1**개의 **구멍**이 존재합니다.


- $\epsilon \in [11,14)$인 동안에는 꼭짓점이 4와 8이 하나의 변으로 이어지고, 앞서 생성되었던 구멍은 매꾸어지며 하나의 연결성분으로 뭉쳐지게 됩니다.
    따라서 Complex에는 **2**개의 **연결성분**이 존재하고 이제는 더 이상 **구멍**은 없습니다.


- 마지막으로 $\epsilon \in [14,\infty)$에는 모든 연결성분들이 뭉쳐지게 됩니다.
    따라서 Complex에는 **1**개의 **연결성분**만 남게 됩니다.

위 과정에서 $\epsilon$이 변해가면서 꼭짓점과 꼭짓점이 변으로 이어지며 하나의 연결성분으로 흡수되고, 또 변들이 모여서 삼각형을 이루며 하나의 연결성분으로 흡수되는 양상을 관찰할 수 있습니다.

즉, 이러한 과정이 계속되면서 하나의 연결성분은 *살아남고* 또 다른 연결성분은 *소멸*합니다.

또 때로는 변들이 이어져서 하나의 구멍이 *탄생*하기도 하고 그것이 다시 매꾸어지면서 사라지기도 합니다.

이렇듯 위상적인 정보를 가진 연결성분과 구멍에 대한 정보는 $\epsilon$의 값에 따라 **탄생(birth)** 하거나 **소멸(death)** 합니다.

이를 수집하면 하나의 위상적 불변량이 언제 탄생했고 언제 소멸했는지, 즉 얼마나 *지속(persist)* 했는지를 알 수 있습니다.

이러한 정보를 $(birth, death)$의 점으로 표시해 나타낸 것을 **Persistent Diagram**이라고 합니다.

다시 말해, Persistent diagram은 $\epsilon$을 변화시켜가며 구축한 Vietoris-Rips complex에서 발생하는 연결성분과 구멍 혹은 그 이상의 차원의 일반화된 구멍이 언제 탄생하고 소멸하는지 점 $(birth, death)$로 표현한 것입니다.



`giotto-tda`는 Vietoris-Rips complex를 통해 얻은 Persistent diagram을 손쉽게 그려주는 기능을 가지고 있습니다.

`giotto-tda`의 작동방식은 `sklearn`과 매우 유사하기 때문에 `sklearn`에 익숙하다면 쉽게 사용할 수 있습니다.

`giotto-tda`를 통해 위 예제의 `point_cloud`에 대한 Persistent diagram을 그려보겠습니다.

1. 먼저 `from gtda.homology`에서 Vietoris-Rips complex로부터 Persistent 정보를 추출하는 `VietorisRipsPersistence`를 임포트합니다. `gtda.homology`에는 Vietoris-Rips 외에도 Cech complex 혹은 이미지 데이터에 주로 사용되는 Cubical complex로부터 Persistence를 얻는 기능들이 수록되어 있습니다. [참고](https://giotto-ai.github.io/gtda-docs/0.5.1/modules/homology.html)

2. `VietorisRipsPersistence`는 대표적으로 `metric`과 `homology_dimensions`라는 파라미터를 가지고 있습니다. `metric`은 Vietoris Rips complex를 구축할 때 사용될 metric의 종류를 선택하는 파라미터이며, 기본적으로 `euclidean` metric이 적용되어 있습니다. 연구주제에 따라 다른 metric을 사용할 수 있습니다. `homology_dimensions`는 complex에서 감지할 호몰로지의 차원을 선택하는 파라미터입니다. 위 예제에서는 연결성분(0차원)과 구멍(1차원)에 대한 정보를 알기 원하므로 `homology_dimensions = [0,1]`로 설정해줍니다. 더 많은 파라미터에 대한 정보는 [여기](https://giotto-ai.github.io/gtda-docs/0.5.1/modules/generated/homology/gtda.homology.VietorisRipsPersistence.html#gtda.homology.VietorisRipsPersistence)서 얻을 수 있습니다.

3. `VietorisRipsPersistence` 변환기를 정의하고 나면, 사이킷런과 유사하게 `fit`으로 포인트클라우드를 학습시키고 `transform`으로 Persistent diagram으로 변환할 수 있습니다. 혹은 `fit_transform`으로 이 과정을 한번에 수행할 수 있습니다. 이 변환기는 여러 개의 포인트 클라우드를 포함하는 리스트를 동시에 변환시켜주는 기능을 가지고 있습니다. 따라서 단일 포인트 클라우드를 변환시킬 경우 `[None, :, :]`의 shape으로 변환해주어야 합니다.

4. `from gtda.plotting`에는 `plotly` 라이브러리를 활용해 포인트 클라우드나 Persistent diagram을 그려줍니다. `plot_diagram`은 Persistent diagram이 저장된 변수를 그림으로 그려줍니다.

```python
## Example. Persistent diagram
from gtda.homology import VietorisRipsPersistence           #Vietoris-Rips complex를 사용해 PD를 얻는 변환기입니다. 
from gtda.plotting import plot_diagram                      #plotly를 이용해 PD를 보여주는 기능입니다.

VR = VietorisRipsPersistence(homology_dimensions = [0,1],   # 먼저 VietorisRipsPersistence 변환기를 정의합니다.
                             reduced_homology=False)        # reduced_homology 변수는 처음부터 끝까지 존재하는 연결성분을 무시할 것인지 정하는 파라미터입니다.

PD = VR.fit_transform(point_cloud[None, :, :])              # 변환기에 point_cloud를 [None, :, :] 쉐입으로 학습/변환 하면 PD를 얻을 수 있습니다.
PD
```

변수 `PD`에 Persistent diagram을 저장했습니다. 

`PD`는 8개의 3차원 `array`로 구성되어 있습니다. 

각 `array`는 [birth time, death time, homology dimension]의 정보로 구성됩니다.

이제 `plot_diagram`을 사용해 Persistent diagram을 시각화해봅시다.

```python
## Plot persistent diagram

plot_diagram(PD[0])

# PD가 아닌 PD[0]을 입력하는 이유는 
# 변환기 VietorisRipsPersistence가 여러 개의 포인트 클라우드를 변환하면 여러 개의 Persistent diagram의 리스트로 저장하기 때문인데,
# 우리는 하나의 포인트 클라우드를 변환했기에, 리스트 PD의 유일한 첫 번째 원소가 실제 Persistent diagram이기 때문입니다.
```

주어진 포인트 클라우드로부터 Persistent diagram을 얻었습니다.

위 정보를 관찰해보면, 8개의 연결성분이 처음부터 존재했었고 각각이 언제 사라졌는지 살펴볼 수 있으며 아주 잠깐동안 구멍(그림에서 연두색점)이 존재했다가 사라졌음을 알 수 있습니다.

일반적으로 Persistent diagram에서 대각선에 가까운 점일수록 지속시간이 짧으므로, 노이즈로 인식되곤 합니다.

결과적으로 우리는 주어진 포인트 클라우드의 위상적인 정보를 정량적인 점들의 모임인 Persistent diagram으로 변환할 수 있었습니다.

이제 이 정보로 우리는 무엇을 할 수 있을까요?

# 4. Classification

여기서는 눈으로는 비슷해보이지만, 실제로는 전혀 다른 분포에서 추출된 포인트 클라우드를 위상적인 정보로 구별하는 실험을 해볼 것입니다.

첫번째 포인트 클라우드는 가우시안 정규분포로부터 추출된 200개의 점으로 구성됩니다.

따라서 통계적으로는 이 포인트 클라우드의 기하학적인 정보는 정규분포 곡면이 가지고 있을 기하학적인 정보와 유사할 것입니다.

두번째 포인트 클라우드는 임의로 선택된 200개의 각을 추출한 뒤, 인접한 두 원을 원을 따라 일정한 노이즈를 적용시켜 흩어둔 정보입니다.

서로 다른 프로세스를 통해 생성된 포인트 클라우드이지만, 눈으로 봐서는 어떤 것이 어떤 프로세스를 통해 만들어 졌는지는 구별하기 어렵습니다.

우리는 여기서 TDA를 통해 이 둘을 구별할 수 있는 패턴을 찾아낼 수 있을지 실험해볼 것입니다.

```python
## Normal distribution으로부터 생성된 Point cloud
import numpy as np
point_cloud1 = np.random.randn(200,2)
point_cloud1.shape
```

```python
## Plot point cloud1
from gtda.plotting import plot_point_cloud

plot_point_cloud(point_cloud1)
```

```python
## Circle 형태를 띈 데이터로부터 생성된 Point cloud
rand_angles = 2*np.pi*np.random.rand(200)       # 랜덤하게 각을 200개 뽑습니다.
rand_noise = np.random.rand(200,1)         # 구간 [0,1]에서 200개의 랜덤한 노이즈를 생성합니다.
circle1 = np.array([np.cos(rand_angles[:100])+0.5,np.sin(rand_angles[:100])]).T    # 중심이 (0.05,0)인 원에서 랜덤하게 100개의 점을 추출합니다.
circle2 = np.array([np.cos(rand_angles[100:])-0.5,np.sin(rand_angles[100:])]).T    # 중심이 (-0.05,0)인 원에서 랜덤하게 100개의 점을 추출합니다.
point_cloud2 = rand_noise*np.vstack((circle1,circle2))  # 위 점들을 모두 합친 뒤, 각각에 노이즈를 적용합니다.
plot_point_cloud(point_cloud2)
```

위 두 포인트 클라우드들이 눈으로 쉽게 구별할 수 있나요?

어떤 것이 어떤 분포를 따르는 것일까요? 

두 포인트 클라우드에 기하학적인 차이점이 존재할까요?

Persistent diagram으로 이들의 기하학적 차이점을 관찰해봅시다.

```python
from gtda.homology import VietorisRipsPersistence

VR = VietorisRipsPersistence(homology_dimensions = [0,1])
point_clouds = [point_cloud1, point_cloud2]  # VietorisRipsPersistence 변환기는 한번에 여러 개의 포인트클라우드를 변환시킬 수 있습니다.
diagrams = VR.fit_transform(point_clouds)    # 이때 diagrams는 PD의 리스트로, 입력한 포인트 클라우드의 순서에 대응하는 PD가 들어있습니다.
diagrams.shape  # 2개의 포인트 클라우드를 입력했기 때문에 2개의 PD를 가지고 있고, 각각은 비교하기 좋게 (n,3) 사이즈로 통일되어 저장됩니다.
```

먼저 첫번째 포인트클라우드의 Persistent diagram을 관찰해봅시다.

```python
from gtda.plotting import plot_diagram
plot_diagram(diagrams[0])
```

다음으로 두번째 포인트클라우드의 Persistent diagram입니다.

```python
plot_diagram(diagrams[1])
```

어떤 차이가 보이시나요? 

두번째 PD에서는 첫번째와 달리 유독 H1(구멍)에 대한 Persistence가 강하게 나타남을 관찰할 수 있습니다.

이는 우연일까요? 이러한 정보를 머신러닝을 통해 학습시키고 분류기로 만들어 테스트해볼 수 있을까요?

일반적으로 Persistent diagram은 머신러닝 학습기의 입력으로 적절하지 않습니다.

지금은 `giotto-tda`의 변환기가 임의로 두 diagram의 사이즈를 맞추어 주었지만, 일반적으로 두 데이터의 diagram이 같은 사이즈를 가질 이유는 없습니다.

이렇게 사이즈가 다를 경우 머신러닝 학습기의 입력으로 사용하는 것에 어려움이 있습니다. 

그렇기 때문에 TDA 연구자들은 이 Persistent diagram의 정보를 충분히 보존하면서 머신러닝의 입력으로 사용하기에 규격화된 여러 Vectorization method들을 개발했습니다.
